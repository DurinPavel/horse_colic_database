Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.


---
Analyze values by columns, calculate basic statistics, find outliers.


```python
# Представляется, что описывать полностью имеет смысл только 3 столбца:
# rectal temperature, pulse и respiratory rate,
# т.к. остальные столбцы - это категории
# -----------------------------------------------------------------------------
# It seems that only 3 columns make sense to describe in full:
# rectal temperature, pulse and respiratory rate,
# because the rest of the columns are categories

df[['rectal temperature', 'pulse', 'respiratory rate']].describe().round(1)
```

```python
# Рассчитаем еще и моду.
# Медиана у нас есть - 50 перцентиль.
# ---------------------------------------------------------------------------
# Let's also calculate the fashion.
# The median we have is the 50th percentile.

df[['rectal temperature', 'pulse', 'respiratory rate']].mode()
```

```python
# Если мы проанализируем все столбцы с категориями, то увидим явные выбросы
# только в столбце Age.
# ----------------------------------------------------------------------------
# If we analyze all columns with categories, we will see explicit outliers
# only in the Age column.

df[(df['Age'] <1 ) | (df['Age'] > 2)].head()
```

```python
# Строк с отклонениями - 24.
# Это довольно много.
# -----------------------------------------------------------------------------
# Lines with deviations - 24.
# This is quite a lot.

len(df[(df['Age'] < 1 ) | (df['Age'] > 2)])
```

```python
# Причина этих выбросов - явно ошибка ввода.
# Мы видим, что в графе Age нет категории 2.
# -----------------------------------------------------------------------------
# The reason for these outliers is clearly an input error.
# We see that there is no category 2 in the Age column.

df[(df['Age'] == 2 )].head()
```

```python
# Можем предположить, что при вводе перепутали цифры 2 и 9.
# Заменим все 9 на 2.
# -----------------------------------------------------------------------------
# We can assume that the numbers 2 and 9 were mixed up when entering.
# Replace all 9 with 2.

df['Age'] = df['Age'].replace(9, 2)
df.head()
```

```python
# Найдем выбросы в анализируемых графах.
# Исходя из того, что показатель respiratory rate изначально имеет сомнительную
# полезность (его описание), и глядя на нашу выбоку, можем предположить,
# что аномальные значения имеются по показателю pulse.
# Найдем выбросы через межквартильный размах, посмотрим, что это за выбросы.
# Мы видим, что повышенные значения пульса характерны исключительно для молодых
# лошадей, а также в выбросы попали пропуски. Их по графе pulse 24.
# Мы не будем удалять выбросы и оставим рассчитанные выше статистики.
# -----------------------------------------------------------------------------
# Find outliers in the analyzed graphs.
# Based on the fact that the respiratory rate initially has a dubious
# utility (its description), and looking at our sample, we can assume
# that there are abnormal values in pulse.
# Find outliers through the interquartile range, see what kind of outliers
# they are.
# We see that increased heart rate values are characteristic only for young
# of horses, as well as omissions were included in the emissions. There are
# 24 of them in the pulse column.
# We will not remove outliers and will keep the statistics calculated above.

q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
df_outliers = df[~df['pulse'].between(lower_bound, upper_bound, inclusive=True)]
df_outliers.head(50)
```

###**Задание 3**


---
###**Task 3**


Рассчитать количество пропусков для всех выбранных столбцов. Принять и  
обосновать решение о методе заполнения пропусков по каждому столбцу на основе  
рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать  
датафрейм, в котором пропуски будут отсутствовать.


---
Calculate the number of gaps for all selected columns. Make and justify the  
decision on the method of filling in the gaps for each column based on the  
calculated statistics and the possible relationship of values in them. Form a  
dataframe in which there will be no gaps.

```python
# Рассчитаем количество пропусков для всех столбцов
# Видим, что полностью пустых строк у нас нет
# -----------------------------------------------------------------------------
# Calculate the number of gaps for all columns
# We see that we do not have completely empty lines

df.isna().sum()
```

```python
# Для начала удалим строки, которые имеют пропуски сразу по 3 графам:
# rectal temperature, pulse, respiratory rate, т.к. включение их в
# в анализ не имеет смысла.
# -----------------------------------------------------------------------------
# First, delete the lines that have gaps in 3 columns at once:
# rectal temperature, pulse, respiratory rate, because including them in
# in the analysis does not make sense.

df_no_full_passes = df.dropna(subset=['rectal temperature', 'pulse', \
                                      'respiratory rate'], how='all')
df_no_full_passes.isna().sum()
```

```python
# Представляется, что корректнее рассчитывать показатели
# в привязке к той или иной категории.
# Рассмотрим показатели для оперированных лошадей.
# -----------------------------------------------------------------------------
# It seems that it is more correct to calculate indicators
# in relation to a particular category.
# Consider indicators for operated horses.

df_surgery = df_no_full_passes[(df_no_full_passes['surgery?'] == 1)]
df_surgery[['rectal temperature', 'pulse', 'respiratory rate']] \
            .describe().round(1)
```

```python
df_surgery[['rectal temperature', 'pulse', 'respiratory rate']].mode()
```

```python
# И показатели для лошадей, которых лечили без операции.
# -----------------------------------------------------------------------------
# And figures for horses treated without surgery.

df_without_surgery = df_no_full_passes[(df_no_full_passes['surgery?'] == 2)]
df_without_surgery[['rectal temperature', 'pulse', 'respiratory rate']] \
                    .describe().round(1)
```

```python
df_without_surgery[['rectal temperature', 'pulse', 'respiratory rate']].mode()
```

```python
# Сделаем копию датафрейма - иначе выходит предупреждение.
# И заменим пропуски в графе rectal temperature на среднее значение
# в связке с параметром surgery? (небольшой разброс)
# В графах pulse и respiratory rate пропуски заменим на медиану
# в связке с параметром surgery? (разброс значительный, медиана посередине между
# средним и модой)
# Пропуски в графах surgery?, temperature of extremities и outcome мы заменим
# модой без привязки к другим параметрам
# Пропуски в графе pain заменим на предыдущие заполненные значения, потому что
# четкой взаимосвязи с какими-либо другими параметрами не прослеживается
# -----------------------------------------------------------------------------
# Make a copy of the dataframe - otherwise a warning will be issued.
# And replace the gaps in the rectal temperature column with the average value
# in conjunction with surgery? (small spread)
# In the pulse and respiratory rate graphs, replace the gaps with the median
# in conjunction with surgery? (significant spread, median in the middle between
# medium and fashion)
# Omissions in the columns surgery?, temperature of extremities and outcome we
# will replace
# fashion without reference to other parameters
# The gaps in the pain column will be replaced by the previous filled values,
# because there is no clear relationship with any other parameters

df_wp = df_no_full_passes.copy()
df_wp['rectal temperature'] = df_wp['rectal temperature'].fillna(
    (df_no_full_passes.groupby(['surgery?'])['rectal temperature']
     .transform('mean'))
    )
df_wp['pulse'] = df_wp['pulse'].fillna(
    (df_no_full_passes.groupby(['surgery?'])['pulse']
     .transform('median'))
    )
df_wp['respiratory rate'] = df_wp['respiratory rate'].fillna(
    (df_no_full_passes.groupby(['surgery?'])['respiratory rate']
     .transform('median'))
    )
df_wp['surgery?'] = df_wp['surgery?'].fillna(
    (df_no_full_passes['surgery?'].mode()[0])
    )
df_wp['outcome'] = df_wp['outcome'].fillna(
    (df_no_full_passes['outcome'].mode()[0])
    )
df_wp['temperature of extremities'] = df_wp['temperature of extremities'].fillna(
    (df_no_full_passes['temperature of extremities'].mode()[0])
    )
df_wp['pain'] = df_wp['pain'].fillna(method='ffill', inplace=False)
```

```python
# Проверим, что у нас с пропусками
# -----------------------------------------------------------------------------
# Check what we have with passes

df_wp.isna().sum()
```
